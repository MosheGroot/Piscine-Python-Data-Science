## Imports

```python
import pandas as pd
```

## Read the data

```python
df = pd.read_json('../data/auto.json',
                  orient='records'
                 )

df.set_index('CarNumber',
             inplace=True
            )

df
```

## Selects
### The rows only where the `Fines` are more than `2,100`

```python
df.loc[df['Fines'] > 2100]
```

###  The rows only where the `Fines` are more than `2,100` and the `Refund` equals `2`

```python
df.loc[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

## The rows only where the `Model`s are from the list: `[’Focus’, ’Corolla’]`

```python
condition_result = df['Model'].apply(lambda x: x in ['Focus', 'Corolla'])
df.loc[condition_result]
```

### The rows only where the `CarNumber` is from the list: `[’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]`

```python
car_numbers_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']

condition_result = map(lambda x: x in car_numbers_list, df.index)

df.loc[car_numbers_list]
```

##  Aggregations with the `Make` and the `Model`
### Display the median fines grouped by the `Make`

```python
df.groupby('Make')['Fines'].agg(['median'])
```

### Display the median fines grouped by the `Make` and the `Model`

```python
fines_grouped_by_make_n_model = df.groupby(['Make', 'Model'])['Fines']
fines_grouped_by_make_n_model.agg(['median'])
```

### Display the number of fines grouped by the `Make` and the `Model`

```python
fines_grouped_by_make_n_model.agg(['count'])
```

### Display the minimum and the maximus fines grouped by the `Make` and the `Model`

```python
fines_grouped_by_make_n_model.agg(['min', 'max'])
```

### Display the standart deviation of the fines grouped by the `Make` and the `Model`

```python
fines_grouped_by_make_n_model.agg(['std'])
```

## Aggregations with the `CarNumber`
### Display the `CarNumber`s grouped by the number of the `Fines` in descending order`

```python
number_of_fines_per_carnumber = df.reset_index()[['CarNumber', 'Fines']].groupby(by=['CarNumber'], as_index=False).count()
number_of_fines_per_carnumber.groupby(by=['Fines', 'CarNumber']).first().iloc[::-1]
```

### Select from initial dataframe all the rows corresponding to the top-1 `CarNumber`


```python
df.loc[df.index[0]]
```

### Display the `CarNumber`s grouped by the sum of the `Fines` in descending order

```python
number_of_fines_per_carnumber = df.reset_index()[['CarNumber', 'Fines']].groupby(by=['CarNumber'], as_index=False).sum()
number_of_fines_per_carnumber.groupby(by=['Fines', 'CarNumber']).first().iloc[::-1]
```
