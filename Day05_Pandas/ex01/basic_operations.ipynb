## Imports

```python
import pandas as pd
```

## Read the data

```python
df = pd.read_csv('../data/feed-views.log',
                      sep='\t',
                      names=('datetime', 'user'),
                      parse_dates=['datetime'])
df
```

```python
df.info()
```

## Split `datetime` to columns

```python
df['year'] = pd.DatetimeIndex(df['datetime']).year
df['month'] = pd.DatetimeIndex(df['datetime']).month
df['day'] = pd.DatetimeIndex(df['datetime']).day
df['hour'] = pd.DatetimeIndex(df['datetime']).hour
df['minute'] = pd.DatetimeIndex(df['datetime']).minute
df['second'] = pd.DatetimeIndex(df['datetime']).second

df
```

## `Daytime` column

```python
df['daytime'] = pd.cut(df['hour'],
                            bins=[-1, 3, 6, 10, 16, 19, 24],
                            labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
                           )
df
```

## Assign the column `user` as the index

```python
df.set_index('user', inplace=True)
df
```

## Counts

```python
df.count(axis='index')
```

```python
df['daytime'].value_counts()
```

## Sort by `hour`, `minute`, and `second` in ascending order

```python
df = df.sort_values(by=['hour', 'minute', 'second'])
df
```

## Calculate the maximum of `hour` for the rows where the time of day is `night`

```python
max_night_hour = (df['hour'].loc[df['daytime'] == 'night']).max()
max_night_hour
```

## Calculate the minimum of `hour` for the rows where the time of day is `morning`

```python
min_morning_hour = (df['hour'].loc[df['daytime'] == 'morning']).min()
min_morning_hour
```

## Find out who visited the page at those hours

```python
df.loc[df['hour'] == max_night_hour]
```

```python
df.loc[df['hour'] == min_morning_hour]
```

## Calculate the mode for the `hour` and `daytime`

```python
df['hour'].mode()
```

```python
df['daytime'].mode()
```

## 3 earliest hours in the morning and the corresponding usernames 


```python
df.loc[df['daytime'] == 'morning'].nsmallest(3, columns=['hour'])['hour']
```

## 3 latest hours and the usernames

```python
df.loc[df['daytime'] == 'morning'].nlargest(3, columns=['hour'])['hour']
```

## Get the basic statistics for the columns

```python
data_describe = df.describe()
data_describe
```

## Interquartile range for the `hour`

```python
iqr = data_describe['hour']['75%'] - data_describe['hour']['25%']
iqr
```
