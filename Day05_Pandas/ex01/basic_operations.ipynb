## Code

### Imports

```python
import pandas as pd
```

### Read data

```python
data_df = pd.read_csv('../data/feed-views.log',
                      sep='\t',
                      names=('datetime', 'user'),
                     )
data_df
```

### Convert datetime column to datetime64[ns] Dtype and split datetime to columns

```python
data_df['datetime'] = pd.to_datetime(data_df['datetime'])

data_df['year'] = pd.DatetimeIndex(data_df['datetime']).year
data_df['month'] = pd.DatetimeIndex(data_df['datetime']).month
data_df['day'] = pd.DatetimeIndex(data_df['datetime']).day
data_df['hour'] = pd.DatetimeIndex(data_df['datetime']).hour
data_df['minute'] = pd.DatetimeIndex(data_df['datetime']).minute
data_df['second'] = pd.DatetimeIndex(data_df['datetime']).second
```

```python
data_df
```

### Daytime column

```python
data_df['daytime'] = pd.cut(data_df['hour'],
                            bins=[-1, 3, 6, 10, 16, 19, 24],
                            labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
                           )
```

```python
data_df
```

### Assign the column user as the index (why?)

```python
data_df.set_index('user', inplace=True)
```

```python
data_df
```

### Counts

```python
data_df.count(axis='index')
```

```python
data_df['daytime'].value_counts()
```

### Sort by hour, minute, and second in ascending order

```python
data_df = data_df.sort_values(by=['hour', 'minute', 'second'])
```

```python
data_df
```

### Calculate the maximum of hour for the rows where the time of day is night

```python
max_night_hour = (data_df['hour'].loc[data_df['daytime'] == 'night']).max()
max_night_hour
```

### Calculate the minimum of hour for the rows where the time of day is morning

```python
min_morning_hour = (data_df['hour'].loc[data_df['daytime'] == 'morning']).min()
min_morning_hour
```

### Find out who visited the page at those hours

```python
data_df.loc[data_df['hour'] == max_night_hour]
```

```python
data_df.loc[data_df['hour'] == min_morning_hour]
```

### Calculate the mode for the hour and daytime

```python
data_df['hour'].mode()
```

```python
data_df['daytime'].mode()
```

### 3 earliest hours in the morning and the corresponding usernames 


```python
data_df.loc[data_df['daytime'] == 'morning'].nsmallest(3, columns=['hour'])['hour']
```

### 3 latest hours and the usernames

```python
data_df.loc[data_df['daytime'] == 'morning'].nlargest(3, columns=['hour'])['hour']
```

### Get the basic statistics for the columns

```python
data_describe = data_df.describe()
data_describe
```

### Interquartile range for the hour

```python
iqr = data_describe['hour']['75%'] - data_describe['hour']['25%']
iqr
```
