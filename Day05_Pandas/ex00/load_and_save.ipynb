## Code

### Imports

```python
import pandas as pd
```

### Read data

```python
data_df = pd.read_csv('../data/feed-views.log',
                      sep='\t',
                      names=('datetime', 'user'),
                      index_col='datetime',
                      skiprows=(2, 3),
                      skipfooter=2,
                      engine='python'
                     )
data_df
```

### Rename the column

```python
data_df.rename_axis('date_time', axis='index', inplace=True)
data_df
```

### Save to csv with delim ';'

```python
data_df.to_csv('feed-views-semicolon.log',
               sep=';'
              )
```
