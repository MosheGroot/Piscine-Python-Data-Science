## Imports

```python
import pandas as pd
```

## Read the data

```python
df = pd.read_csv('../data/feed-views.log',
                      sep='\t',
                      names=('datetime', 'user'),
                      index_col='datetime',
                      skiprows=(2, 3),
                      skipfooter=2,
                      engine='python'
                     )
df
```

## Rename the column

```python
df.rename_axis('date_time', axis='index', inplace=True)
df
```

## Save to csv with delim ';'

```python
df.to_csv('../data/feed-views-semicolon.log',
          sep=';'
         )
```
