## Imports

```python
import pandas as pd
import sqlite3
```

## Connect to database

```python
db_connection = sqlite3.connect('../data/checking-logs.sqlite')
```

## Get all commit datetimes from `checker`

```python
sql_query = """
SELECT
    timestamp as datetime
FROM 
    checker
WHERE
    uid LIKE 'user_%'
"""

commits_datetimes_df = pd.read_sql(sql_query, db_connection, parse_dates=['datetime'])
commits_datetimes_df
```

```python
commits_datetimes_df.info()
```

## Get daytime and daytime counts per date

```python
commits_datetimes_df['daytime'] = pd.cut(commits_datetimes_df['datetime'].dt.hour,
                                         bins=[-1, 4, 10, 17, 24],
                                         labels=['night', 'morning', 'afternoon', 'evening']
                                        )
commits_datetimes_df
```

```python
daytime_counts_per_day_df = commits_datetimes_df.groupby(by=[commits_datetimes_df['datetime'].dt.date, commits_datetimes_df['daytime']]).count()
daytime_counts_per_day_df.rename(columns={'datetime': 'count'}, inplace=True)
daytime_counts_per_day_df = daytime_counts_per_day_df.unstack().droplevel(0, axis='columns')
daytime_counts_per_day_df
```

## Create a graph

```python
import matplotlib
```

```python
plt = daytime_counts_per_day_df.plot.bar(figsize=(15, 8),
                                      fontsize=8,
                                      title='Commits per day',
                                      rot=90,
                                      alpha=0.5,
                                      stacked=True
                                     )
plt.grid(axis='y')
plt
```

## Disconnect database

```python
db_connection.close()
```
