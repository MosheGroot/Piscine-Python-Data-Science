## Imports

```python
import pandas as pd
```

## Load the data

```python
ab_testing_df = pd.read_csv('../data/ab-test.csv')
ab_testing_df
```

```python
ab_testing_df.info()
```

## Prepare the data: split to "after" and "before" first newsfeed view

```python
data_before = ab_testing_df[ab_testing_df['time'] == 'before'][['group', 'diff']]
data_before = [data_before[data_before['group'] == group]['diff'].tolist() for group in ('control', 'test')]
data_before
```

```python
data_after = ab_testing_df[ab_testing_df['time'] == 'after'][['group', 'diff']]
data_after = [data_after[data_after['group'] == group]['diff'].tolist() for group in ('control', 'test')]
data_after
```

## Create a graph

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 8))

axs[0].boxplot(data_after,
               labels=['control', 'test'],
               showmeans=True,
               boxprops={ 'color': 'green', 'linewidth': 3 },
               whiskerprops={ 'color': 'red'},
               capprops={ 'color': 'red' },
               medianprops={ 'color': 'red', 'linewidth': 2 }
              )
axs[0].set_title('after')
axs[0].grid('xy')

axs[1].boxplot(data_before, 
               labels=['control', 'test'],
               showmeans=True,
               boxprops={ 'color': 'green', 'linewidth': 3 },
               whiskerprops={ 'color': 'red' },
               capprops={ 'color': 'red' },
               medianprops={ 'color': 'red', 'linewidth': 2 }
              )
axs[1].set_title('before')
axs[1].grid('xy')

plt.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```
